```python
quarter = 4
females = ["F_P101", "F_P102", "F_P103", "F_P104", "F_P105"]

male_file_name = "C:/Users/jvkim/Downloads/2025_2026 KFH Spawn Matrix.xlsx - Match Maker.csv"
male_use = pd.read_csv(male_file_name).iloc[398:500, 1:3].rename(columns = {"Males:": "Male", "Unnamed: 2" : "Uses"})
male_use["Uses"] = male_use["Uses"].astype(int)
male_use
```

Run the solver. The number of avaliable solutions are shown, as well as the table for the top solution based on the spawn matrix ranking
